```python
import imageio

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import cv2
```

```python
img_rgb = imageio.imread('toad_and_jiang.jpg')
```

```python
img = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
```

```python
toad = img[:,:140]
```

```python
plt.imshow(toad, cmap='gray')
```

```python
moments = cv2.moments(toad)
```

```python
huMoments = cv2.HuMoments(moments)
```

```python
moments
```

```python
huMoments
```

```python
np.log(huMoments)
```

```python
plt.imshow(toad[:toad.shape[0]//3,:toad.shape[1]//3])
```

```python
moments = cv2.moments(toad[:toad.shape[0]//3,:toad.shape[1]//3])
```

```python
moments
```

```python
huMoments = cv2.HuMoments(moments)
```
