```julia
X = vec(convert(Array{Float64}, toad_r))

_size = 100
X_seg=1:1:100
N_seg = zeros(Int,_size)
for Xi in X
    #println(Int(trunc(Xi*(_size-1))))
    N_seg[Int(trunc(Xi*(_size-1)))+1] += 1
end
```

```julia
bar(N_seg)
```

```julia
Y_seg = sqrt(100/size(X)[1]) .* sqrt.(N_seg .+ 0.25);
```

```julia
plot(X_seg, Y_seg)
```

```julia
h = 1.0

L_n = K.((reshape(X_seg, size(X_seg,1),1) .- reshape(X_seg, 1, size(X_seg,1)))./h)
```

```julia
L = L_n ./ reshape(sum(L_n,dims=2),size(L_n,1),1)
```

```julia
sum(L,dims=2)[1]
```

```julia
plot(X_seg, L * Y_seg)
```

```julia
n = size(X_seg,1)
nu = tr(L)
nu_tilde = tr(transpose(L) * L)
```

```julia
sigma2_seg = sum((Y_seg .- L * Y_seg).^2)/(n-2*nu+nu_tilde)
```

```julia
Vr = sigma2_seg .* sum(L.^2,dims=2)
```

```julia
Sr = sqrt.(Vr)
```

```julia
plot(X_seg ./ 100, L * Y_seg, ribbon=Sr.*3)
```

```julia
X = vec(convert(Array{Float64}, toad_b))

_size = 100
X_seg=1:1:100
N_seg = zeros(Int,_size)
for Xi in X
    #println(Int(trunc(Xi*(_size-1))))
    N_seg[Int(trunc(Xi*(_size-1)))+1] += 1
end

bar(N_seg)
```

```julia
Y_seg = sqrt(100/size(X)[1]) .* sqrt.(N_seg .+ 0.25);
```

```julia
plot(X_seg, Y_seg)
```

```julia
h = 1.0

L_n = K.((reshape(X_seg, size(X_seg,1),1) .- reshape(X_seg, 1, size(X_seg,1)))./h)
L = L_n ./ reshape(sum(L_n,dims=2),size(L_n,1),1)
```

```julia
n = size(X_seg,1)
nu = tr(L)
nu_tilde = tr(transpose(L) * L)

sigma2_seg = sum((Y_seg .- L * Y_seg).^2)/(n-2*nu+nu_tilde)
```

```julia
Vr = sigma2_seg .* sum(L.^2,dims=2)
```

```julia
Sr = sqrt.(Vr)

plot(X_seg ./ 100, L * Y_seg, ribbon=Sr.*3)
```
