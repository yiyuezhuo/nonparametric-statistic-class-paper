```python
import imageio

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

为了和非参数图像模型做比较，这里产生一些参数图像模型的例子，其实就是一些“有规律的图案”，虽然卷积神经网络如果看成参数模型的话当然可以产生出
不那么贫乏的结果。

```python
_x = np.arange(0,200)
_y = np.arange(0,200)
x,y = np.meshgrid(_x,_y)
```

```python
plt.imshow(x+y, cmap='gray')
```

```python
plt.imshow(np.log(1+x)+np.log(1+y), cmap='gray')
```

```python
plt.imshow(np.sin(1+x)+np.sin(1+y), cmap='gray')
```

```python
plt.imshow(np.sin(0.1*x)+np.sin(0.1*y), cmap='gray')
```
